## Business Understanding

### 1. Latar Belakang
Berdasarkan data World Health Organization (WHO), Penyakit Kardiovaskular menempati urutan pertama sebagai penyebab kematian global. Salah satu faktor risiko dominan dari penyakit ini adalah hipertensi dan gaya hidup tidak sehat, kondisi yang seringkali tidak terdiagnosis secara dini karena minimnya gejala klinis yang spesifik.

Meskipun penggunaan alat pengukur tekanan darah mandiri (sphygmomanometer) semakin umum di masyarakat, pemanfaatannya seringkali terbatas pada pemantauan sesaat (snapshot) tanpa disertai analisis tren jangka panjang dan faktor gaya hidup penyerta. Ketiadaan sistem pencatatan cerdas yang terintegrasi mengakibatkan data historis kesehatan pasien tidak dimanfaatkan secara optimal untuk mendeteksi potensi komplikasi.

Oleh karena itu, proyek ini mengembangkan CardioGuard, sebuah sistem monitoring kesehatan cerdas yang mengimplementasikan pendekatan Hybrid Data Mining. Dengan mengintegrasikan algoritma K-Means Clustering untuk segmentasi profil risiko dan Random Forest Classification untuk prediksi probabilitas, sistem ini dirancang untuk memberikan wawasan prediktif yang akurat guna mendukung upaya pencegahan medis secara dini dan mandiri.

### 2. Tujuan
Proyek ini bertujuan untuk merancang dan membangun pipeline Machine Learning end-to-end untuk aplikasi CardioGuard. Secara spesifik, tujuan penelitian ini meliputi:

* **Implementasi Hybrid Data Mining:** Mengembangkan model analisis data yang menggabungkan metode Unsupervised Learning (Klastering) dan Supervised Learning (Klasifikasi) untuk memetakan risiko kesehatan secara komprehensif.

* **Segmentasi Profil Pasien (Patient Stratification):** Menerapkan algoritma K-Means Clustering untuk mengelompokkan data pasien berdasarkan kemiripan karakteristik klinis (usia, IMT, tekanan darah) dan gaya hidup (kolesterol, glukosa, merokok), guna mengidentifikasi pola risiko yang tersembunyi.

* **Prediksi Risiko Kardiovaskular:** Membangun model klasifikasi Random Forest untuk memprediksi probabilitas keberadaan penyakit kardiovaskular berdasarkan parameter input, dengan fokus pada sensitivitas deteksi.

* **Integrasi Platform (Deployment):** Mengekspor model menjadi artefak siap pakai (.pkl) untuk diintegrasikan ke dalam Backend API Python, yang nantinya akan melayani antarmuka web interaktif CardioGuard untuk keperluan monitoring dan pelaporan historis.


### 3. Metrik Keberhasilan
Keberhasilan pengembangan pipeline pemodelan ini diukur berdasarkan indikator kuantitatif dan kualitatif sebagai berikut:

**A. Kinerja Model (Model Performance Metrics)**

 * **Akurasi Klasifikasi (Accuracy):** Model Random Forest diharapkan mencapai tingkat akurasi minimal 70% pada data pengujian (test set).

 * **Sensitivitas (Recall):** Memprioritaskan nilai Recall yang tinggi (> 75%) untuk meminimalkan False Negative. Dalam konteks medis, kegagalan mendeteksi pasien yang sebenarnya berisiko (False Negative) jauh lebih fatal daripada kesalahan mendeteksi orang sehat sebagai berisiko (False Positive).

 * **Kualitas Klaster (Clustering Quality):** Evaluasi hasil segmentasi K-Means menggunakan Silhouette Score dengan nilai positif, mengindikasikan bahwa setiap klaster terpisah dengan baik (well-separated).

**B. Kesiapan Sistem (System Functionality)**

 * **Efisiensi Inferensi:** Model yang diekspor mampu memproses input data pengguna baru dan mengembalikan hasil prediksi dalam waktu respons yang sangat cepat (< 5 detik) saat diintegrasikan ke dalam API.

 * **Ketahanan Skala (Robustness):** Pipeline prapemrosesan (melalui StandardScaler) mampu menangani dan menstandardisasi variasi input data dunia nyata secara konsisten.

### Visualisasi

#### Histplot

```python
sns.set_style("whitegrid")
sns.set_palette("pastel")

num_cols = df_u.select_dtypes(include=['int64', 'float64']).columns

n_cols = 3
n_rows = (len(num_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(16, 4 * n_rows))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    if df_u[col].nunique() < 10:
        sns.countplot(
            data=df_u,
            x=col,
            ax=axes[i],
            color='lightsteelblue'
        )
    else:
        sns.histplot(
            data=df_u,
            x=col,
            bins=25,
            kde=True, 
            ax=axes[i]
        )
        
    axes[i].set_title(f"Distribusi {col}", fontsize=12)
    axes[i].set_xlabel('')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

**Penjelasan:** Melakukan visualisasi histogram plot (histplot) pada seluruh fitur dataset untuk memahami distribusi demografis, metrik klinis, dan faktor gaya hidup pasien. Analisis ini sangat krusial untuk menentukan strategi Data Preparation sebelum data diproses oleh pipeline Machine Learning CardioGuard.

1. **Distribusi id (Identifier)**
Menunjukkan pola seragam (uniform) yang rata dari awal hingga akhir. Hal ini mengonfirmasi bahwa fitur ini murni merupakan penanda urutan unik pasien dan tidak memiliki nilai prediktif.

    * **Tindak Lanjut:** Fitur ini akan dihapus (drop) dari dataset agar tidak menjadi noise pada model prediktif.

3. **Distribusi age (Usia)**
Rentang nilai berada di kisaran puluhan ribu, menandakan metrik usia masih tercatat dalam satuan hari. Distribusi memusat pada rentang usia paruh baya hingga lansia.

    * **Tindak Lanjut:** Diperlukan transformasi matematis (pembagian dengan 365) untuk mengonversi satuan hari menjadi tahun (years). Hal ini diwajibkan agar format data selaras dengan input form profil fisik pada antarmuka aplikasi CardioGuard.

4. **Distribusi gender (Jenis Kelamin)**
Bersifat diskrit dengan dua batang terpisah (1 dan 2), mengonfirmasi format data kategorikal biner yang mewakili Pria dan Wanita.

5. **Distribusi height (Tinggi Badan)**
Menunjukkan pola yang sangat mendekati distribusi normal (bell-shaped) dengan puncak populasi di rentang 160-170 cm. Distribusi yang stabil ini menandakan variasi tinggi badan berada dalam batas wajar. Data ini akan sangat berguna untuk kalkulasi metrik turunan seperti Body Mass Index (BMI).

6. **Distribusi weight (Berat Badan)**
Menyerupai distribusi normal namun sedikit melenceng ke kanan (right-skewed). Puncak data berada di kisaran 60-70 kg, namun terdapat ekor panjang ke arah nilai besar. Ini menandakan keberadaan subset populasi dengan indikasi kelebihan berat badan (obesitas) yang merupakan faktor risiko relevan untuk klastering kardiovaskular.

7. **Distribusi ap_hi (Tekanan Sistolik) dan ap_lo (Tekanan Diastolik)**
Menunjukkan kemencengan ekstrem (extreme skewness) dengan grafik yang secara visual tidak wajar. Data terkonsentrasi di nilai kecil, namun sumbu-x memanjang hingga ribuan (misal: input 12000 yang seharusnya 120 mmHg).

    * **Tindak Lanjut:** Ini adalah anomali data (kemungkinan human error saat penginputan) yang sangat fatal. Diperlukan metode filtering ketat untuk membuang nilai tekanan darah yang berada di luar ambang batas fisiologis manusia agar tidak merusak perhitungan jarak pada model K-Means.

8. **Distribusi cholesterol dan gluc (Glukosa)**
Bersifat ordinal dengan tiga kategori (1: Normal, 2: Di atas Normal, 3: Sangat Tinggi). Grafik didominasi kuat oleh nilai 1 (Normal). Pada aplikasi CardioGuard, fitur ini akan diimplementasikan sebagai parameter gaya hidup (lifestyle) opsional yang secara bawaan (default) disetel pada nilai normal.

9. **Distribusi smoke, alco, dan active (Faktor Gaya Hidup Biner)**
Merupakan fitur biner (0 dan 1). Mayoritas populasi sampel melaporkan profil yang sehat: tidak merokok, tidak mengonsumsi alkohol berlebih, dan aktif secara fisik.

10. **Distribusi cardio (Variabel Target)**
Menunjukkan dua batang yang proporsinya hampir sejajar antara kelas 0 (Sehat/Aman) dan kelas 1 (Sakit/Berisiko).

    * **Tindak Lanjut:** Ini adalah kondisi yang sangat ideal (balanced dataset). Proporsi yang seimbang ini memastikan bahwa model Random Forest tidak akan mengalami bias atau kecenderungan memprediksi kelas mayoritas saja.

**Kesimpulan:** Secara keseluruhan, struktur target prediktif (cardio) sangat ideal untuk algoritma klasifikasi karena tidak mengalami class imbalance. Namun, anomali ekstrem pada metrik tanda vital (ap_hi dan ap_lo) serta ketidaksesuaian format metrik usia (age) menuntut tahapan Data Cleaning yang agresif. Membersihkan, menstandardisasi (menggunakan StandardScaler), dan memastikan data masuk akal secara medis adalah syarat mutlak sebelum dataset ini aman untuk diekstraksi polanya menggunakan algoritma K-Means Clustering dan Random Forest.

#### Boxplot

```python
sns.set_style("whitegrid")
sns.set_palette("pastel")

all_num_cols = df_u.select_dtypes(include=['int64', 'float64']).columns
continuous_cols = [col for col in all_num_cols if df_u[col].nunique() > 10 and col != 'id']

n_cols = 3
n_rows = (len(continuous_cols) + n_cols - 1) // n_cols

fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(16, 4 * n_rows))
axes = axes.flatten()

for i, col in enumerate(continuous_cols):
    sns.boxplot(
        data=df_u,
        x=col,
        ax=axes[i],
        color='lightsteelblue'
    )
    axes[i].set_title(f"Boxplot {col}", fontsize=12)

for j in range(len(continuous_cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

**Penjelasan:** Analisis Boxplot ini diisolasi secara khusus pada fitur pengukuran fisik dan vital yang bersifat kontinu (age, height, weight, ap_hi, ap_lo). Fitur berskala kategorikal dan biner sengaja dieksklusi karena analisis kuartil (Interquartile Range/IQR) tidak relevan untuk data diskrit. Visualisasi ini bertujuan secara spesifik untuk mendeteksi outlier (pencilan) yang berpotensi merusak kalkulasi jarak Euclidean pada algoritma K-Means Clustering CardioGuard.

Berdasarkan kelima visualisasi di atas, diperoleh temuan kritis sebagai berikut:

1. **Boxplot age (Usia):** Memperlihatkan distribusi usia (masih dalam satuan hari) dengan kuartil yang terkonsentrasi pada nilai tinggi, menandakan mayoritas sampel adalah populasi paruh baya hingga lanjut usia. Terdapat sekumpulan outlier di sisi kiri (usia muda).

    * **Tindak Lanjut:** Mengonfirmasi kembali urgensi konversi ke satuan tahun (years) agar data latih selaras dengan format form input antarmuka CardioGuard.

2. **Boxplot height (Tinggi Badan):** Nilai median terpusat secara normal di kisaran 160-165 cm. Namun, ditemukan outlier ekstrem di kedua ujung distribusi (sangat pendek di bawah 100 cm, dan sangat tinggi di atas 200 cm).

    * **Tindak Lanjut:** Titik ekstrem ini dipastikan sebagai measurement error. Pembersihan nilai ini diwajibkan agar fitur Auto-BMI Calculator pada sistem CardioGuard tidak menghasilkan metrik yang bias.

3. **Boxplot weight (Berat Badan):** Memiliki sebaran yang right-skewed (miring ke kanan) dengan median di kisaran 70 kg. Rentetan outlier yang cukup padat di sisi kanan merepresentasikan sub-populasi dengan kondisi berat badan berlebih (obesitas).

    * **Tindak Lanjut:** Dalam konteks HealthTech, nilai ini sebagian besar adalah sinyal klinis yang valid dan akan dipertahankan dalam batas logis medis, karena merupakan metrik pembentuk klaster risiko yang sangat berharga.

4. **Boxplot ap_hi (Tekanan Sistolik):** Anomali Fatal. Rentang kotak utama (IQR) terkompresi menjadi satu garis tipis akibat tertarik oleh outlier ekstrem yang mencapai angka di atas 16.000 mmHg. Nilai ini mustahil secara fisiologis manusia (batas rekor medis biasanya tidak melebihi 300 mmHg) dan murni merupakan noise atau typo.

5. **Boxplot ap_lo (Tekanan Diastolik):** Memperlihatkan pola distorsi yang serupa dengan ap_hi, dengan outlier ekstrem melampaui 10.000 mmHg. Hal ini mengonfirmasi terjadinya kesalahan input pencatatan secara masif pada instrumen pengukuran awal.

**Kesimpulan:** Visualisasi Boxplot ini membuktikan bahwa dataset mentah ini belum siap untuk masuk ke tahap pemodelan. Sementara fitur usia dan berat badan memiliki variasi yang masih dapat ditoleransi, fitur ap_hi dan ap_lo terbukti mengandung data "sampah" akibat human error.

Mengingat arsitektur CardioGuard bertumpu pada K-Means Clustering (yang kalkulasinya sangat rentan terdistorsi oleh titik data ekstrem), penerapan Physiological Threshold Filtering (penyaringan berdasarkan ambang batas medis logis) adalah langkah mutlak. Rentang tekanan darah yang tidak masuk akal harus dihapus sebelum data distandardisasi dan dimasukkan ke dalam pipeline Machine Learning.

#### Pieplot

```python
sns.set_style("whitegrid")

data_cardio = df_u["cardio"].value_counts().sort_index()

colors = sns.color_palette("crest", n_colors=len(data_cardio))

plt.figure(figsize=(6, 6)) 

def func(pct):
    return f"{pct:.1f}%"

wedges, texts, autotexts = plt.pie(
    data_cardio,
    labels=["0 (Sehat / Tidak Berpotensi)", "1 (Sakit / Berpotensi)"],
    autopct=func,
    startangle=90,
    colors=colors,
    explode=(0, 0.05)
)

for autotext in autotexts:
    autotext.set_color("white")
    autotext.set_fontsize(12)
    autotext.set_weight('bold')

plt.title("Distribusi Proporsi Kelas Penyakit Kardiovaskular", fontsize=14, pad=20)
plt.show()
```

**Penjelasan** Melakukan visualisasi komparatif menggunakan Pie Chart (diagram lingkaran) untuk mengonfirmasi persentase proporsi secara absolut pada variabel target klasifikasi, yaitu cardio (Status Risiko Kardiovaskular). Visualisasi ini berfungsi sebagai validasi akhir sebelum data diputuskan layak masuk ke tahap Data Preparation.

Berdasarkan persentase pada visualisasi Pie Chart di atas, diperoleh temuan definitif sebagai berikut:

 * **Simetri Visual:** Proporsi data terbagi secara simetris sempurna, di mana tepat 50.0% dari total populasi dataset merepresentasikan pasien pada kelas 0 (Aman / Risiko Rendah).

 * **Ekuivalensi Risiko:** Sebesar 50.0% sisanya merupakan representasi akurat dari populasi pasien pada kelas 1 (Berisiko Tinggi Kardiovaskular).

 * **Absennya Dominasi Kelas:** Pembagian juring (slice) yang sama besar ini memberikan konfirmasi visual mutlak bahwa tidak terdapat bias demografis atau dominasi kelas (class dominance) dari pihak mana pun di dalam dataset historis ini.

**Kesimpulan:** Visualisasi Pie Chart ini mempertegas temuan Countplot sebelumnya bahwa dataset berada dalam kondisi "Skenario Emas" (Perfectly Balanced).

Dalam pengembangan sistem berbasis Machine Learning seperti CardioGuard, kesetaraan persentase ini memastikan bahwa algoritma klasifikasi (Random Forest) tidak akan memiliki kecenderungan (bias) untuk lebih "menghafal" satu kelas dibandingkan kelas lainnya. Sebagai keputusan pipeline, data ini dinyatakan sangat siap dan ideal untuk masuk ke tahap pemodelan tanpa membebani komputasi dengan teknik resampling tambahan. **Penjelasan:** Melakukan analisis statistik deskriptif univariat (describe()) untuk mengekstraksi ringkasan numerik dari dataset historis. Tahapan ini tidak sekadar untuk melihat rata-rata atau sebaran, melainkan bertindak sebagai Sanity Check (Uji Kelayakan Medis) untuk mengidentifikasi anomali kritis yang berpotensi merusak pipeline Machine Learning CardioGuard.

Berdasarkan tabel metrik statistik di atas, diperoleh temuan dan arahan teknis sebagai berikut:

1. **Kelengkapan Data (Data Completeness)**

    * Seluruh fitur memiliki jumlah observasi (count) yang ekuivalen, yakni 70.000 baris.

    * **Tindak Lanjut:** Hal ini memastikan tidak ada nilai kosong (missing values / NaN) pada tahap ekstraksi awal. Arsitektur pemrosesan tidak perlu membuang beban komputasi untuk modul imputasi data.

2. **Anomali Ekstrem Metrik Vital (Blood Pressure Outliers)**

    * Tensi Sistolik (ap_hi) & Diastolik (ap_lo): Ditemukan distorsi fatal. Nilai minimum tercatat negatif (-150 mmHg) dan maksimum mencapai angka absurd (16.020 mmHg).

    * **Tindak Lanjut:** Secara fisiologis, tekanan darah tidak mungkin bernilai negatif atau melampaui batas anatomis kelangsungan hidup manusia. Ini adalah indikasi kuat terjadinya kesalahan input (human error/typo). Mengingat algoritma K-Means Clustering CardioGuard bergantung pada perhitungan jarak Euclidean, data fiktif ini akan mengacaukan konstelasi klaster. Nilai-nilai ini wajib dieliminasi.

3. **Inkonsistensi Ukuran Fisik (Physical Measurement Errors)**

    * Variabel weight (Berat Badan) memiliki rekor minimum 10.0 kg dan height (Tinggi Badan) 55 cm.

    * **Tindak Lanjut:** Mengingat target pengguna CardioGuard adalah manusia dewasa, angka-angka ini merepresentasikan ukuran bayi atau kesalahan alat ukur. Penerapan filter batas bawah (lower bound filter) sangat diperlukan agar kalkulasi Auto-BMI di aplikasi nanti tidak menghasilkan output yang cacat logika.

4. **Karakteristik Demografi Usia (Age Demographics)**

    * Rata-rata usia populasi berada di angka 19.468 hari (~53 tahun), membentang dari 10.798 hari (29 tahun) hingga 23.713 hari (65 tahun).

    * **Tindak Lanjut:** Data mengonfirmasi bahwa dataset didominasi oleh populasi usia produktif lanjut hingga lansia. Transformasi komputasional dari hari ke tahun (years) wajib dilakukan agar selaras dengan parameter input antarmuka Frontend CardioGuard.

5. **Validasi Matematis Keseimbangan Kelas (Target Balance Confirmation)**

    * Nilai rata-rata (mean) pada variabel target cardio berada di angka 0.4997.

    * **Tindak Lanjut:** Karena rentang nilai target adalah biner (0 dan 1), angka yang nyaris tepat di 0.50 ini membuktikan secara matematis bahwa proporsi kelas Sehat dan Berisiko benar-benar berada di rasio 50:50 (Perfectly Balanced).
  
**Kesimpulan:** Analisis deskriptif ini menyingkap realitas kritis: dataset memiliki tingkat kelengkapan 100%, namun menderita kualitas kelayakan medis (medical validity) yang sangat rendah pada fitur parameter vital dan fisik.

Oleh karena itu, tahapan Data Preparation dalam pipeline CardioGuard tidak akan berfokus pada pengisian data kosong, melainkan dieksekusi dengan pendekatan Physiological Threshold Filtering (Penyaringan Berdasarkan Ambang Batas Medis). Baris data yang mengandung nilai tekanan darah dan ukuran fisik artifisial akan diamputasi tanpa kompromi, guna memastikan algoritma pemodelan hanya mempelajari pola risiko kardiovaskular yang otentik dan akurat.

Algoritma K-Means Clustering akan sepenuhnya rusak dan bias karena titik pusat klasternya (centroid) akan ditarik paksa oleh nilai belasan ribu ini. Temuan ini menjadi mandat mutlak untuk mengeksekusi Physiological Threshold Filtering—memotong data secara ketat berdasarkan ambang batas kelangsungan hidup manusia yang logis—guna memulihkan integritas ruang data sebelum Hybrid Machine Learning dieksekusi.

### Penanganan Outlier

```python
import numpy as np

print("--- STATUS DATA AWAL ---")
print(f"Jumlah baris awal (df_p): {df_p.shape[0]}")

df_p.drop_duplicates(subset=df_p.columns.difference(['id']), keep='first', inplace=True)
print(f"Jumlah baris setelah duplikat dihapus: {df_p.shape[0]}")

df_p['umur_tahun'] = np.floor(df_p['umur'] / 365.25).astype(int)
df_p.drop('umur', axis=1, inplace=True)

kondisi_tensi = (df_p['sistolik'] >= 70) & (df_p['sistolik'] <= 250) & \
                (df_p['diastolik'] >= 40) & (df_p['diastolik'] <= 150) & \
                (df_p['sistolik'] > df_p['diastolik'])

kondisi_fisik = (df_p['tinggi_badan'] >= 100) & (df_p['tinggi_badan'] <= 250) & \
                (df_p['berat_badan'] >= 30) & (df_p['berat_badan'] <= 300)

df_clean = df_p[kondisi_tensi & kondisi_fisik].copy()

cols = ['id', 'umur_tahun'] + [c for c in df_clean.columns if c not in ['id', 'umur_tahun']]
df_clean = df_clean[cols]

print("\n--- HASIL AKHIR DATA PREPROCESSING 1 ---")
print(f"Jumlah baris data bersih (df_clean): {df_clean.shape[0]}")
print(f"Total data sampah (outlier/duplikat) yang dibuang: {df_p.shape[0] - df_clean.shape[0]} baris")

display(df_clean.head())
```

### Ringkasan Hasil Tahapan Data Preprocessing 1

Berdasarkan hasil eksekusi pipeline pada tahapan Data Preparation 1, struktur dan volume dataset historis CardioGuard telah melalui transformasi krusial guna memenuhi standar production-ready. Berikut adalah rincian perubahannya:

1. **Eliminasi Redundansi Data (Data Deduplication):** Sistem pendeteksi otomatis berhasil mengidentifikasi dan mengeliminasi 24 observasi rekam medis yang terekam ganda (menyusutkan basis data dari 70.000 menjadi 69.976 baris). Langkah preventif ini memastikan bahwa setiap baris mewakili entitas pasien yang unik secara komputasi, sehingga mencegah penarikan bobot ganda (double-counting) yang dapat membiasakan perhitungan centroid pada algoritma K-Means Clustering.

2. **Eksekusi Physiological Threshold Filtering:** Implementasi seleksi berbasis batasan logis medis (Domain Knowledge) telah sukses mengamputasi noise ekstrem secara permanen, seperti rekam tekanan darah negatif atau ukuran fisik yang mustahil secara anatomis. Secara kumulatif, total data "sampah" (gabungan duplikat dan outlier medis) yang direduksi berjumlah 1.366 observasi. Pembersihan ini menyelamatkan Random Forest dari risiko mempelajari aturan keputusan (decision rules) yang cacat logika.

3. **Tingkat Retensi Korpus Data yang Optimal (High Data Retention):** Dataset final kini terdiri dari 68.610 observasi yang 100% valid secara klinis. Penyusutan data yang terjadi sangatlah marjinal, yakni hanya berkisar di angka 1,95%. Hal ini membuktikan bahwa lebih dari 98% integritas informasi populasi awal berhasil dipertahankan. Persentase retensi yang masif ini adalah skenario sangat ideal untuk menjaga stabilitas performa dan metrik sensitivitas (recall) pada model prediktif yang akan dibangun.

4. **Standardisasi Skema API & Sinkronisasi Frontend (UI/UX):** Transformasi format dan schema mapping telah dieksekusi untuk menjembatani mesin AI dengan antarmuka aplikasi. Atribut usia telah dikonversi secara matematis dari satuan hari menjadi tahun bulat (umur_tahun). Selanjutnya, translasi seluruh atribut parameter klinis ke dalam Bahasa Indonesia (seperti sistolik, diastolik, tinggi_badan, berat_badan) membuat skema database ini selaras sempurna dengan input form pada Frontend aplikasi CardioGuard (berbasis Lovable.dev). Penyesuaian arsitektur ini menjamin aliran pertukaran data (data payload) yang mulus dari pengguna akhir (end-user) menuju Backend API Python secara real-time.

## Data Preprocessing 2

### Splitting Data

```python
from sklearn.model_selection import train_test_split

print("--- MEMULAI DATA PREPROCESSING 2 ---")
X = df_clean.drop(['id', 'kardiovaskular'], axis=1)
y = df_clean['kardiovaskular']
print("Langkah 1 Selesai: Fitur (X) dan Target (y) berhasil dipisahkan.")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2, 
    stratify=y, 
    random_state=42
)

print("\nLangkah 2 Selesai: Data berhasil dibagi.")
print(f"-> Jumlah Data Latih (X_train): {X_train.shape[0]} baris")
print(f"-> Jumlah Data Uji (X_test): {X_test.shape[0]} baris")

display(X_train.head())
```

**Penjelasan:** Melakukan partisi struktural pada dataset bersih menjadi matriks fitur (independent variables) dan vektor target (dependent variable), dilanjutkan dengan pembagian proporsi Data Latih dan Data Uji. Tahapan ini adalah fondasi isolasi yang krusial sebelum data memasuki fase rekayasa fitur hibrida (K-Means) dan pelatihan model klasifikasi (Random Forest) pada mesin CardioGuard.

Berdasarkan output eksekusi fungsi splitting, arsitektur dataset telah dipersiapkan dengan rincian teknis sebagai berikut:

1. **Isolasi Matriks Fitur & Eliminasi Noise ID:** Seluruh atribut prediktor medis (seperti sistolik, umur_tahun, kolesterol) berhasil diisolasi ke dalam matriks X, sementara label diagnostik aktual (kardiovaskular) dikunci ke dalam vektor y. Atribut id telah diamputasi secara permanen dari pipeline karena murni merupakan indeks sistem (metadata) tanpa bobot prediktif.

    * **Implikasi Sistem:** Pembuangan id ini adalah mandat mutlak; jika dipertahankan, angka urutan acak tersebut akan mendistorsi kalkulasi jarak Euclidean pada algoritma K-Means. Kini, matriks X secara arsitektural telah identik 100% dengan format payload input (JSON) yang nantinya akan dikirimkan oleh aplikasi Frontend CardioGuard.

2. **Distribusi Proporsi Standar Industri (80:20):** Dataset bersih bervolume 68.610 baris dipartisi menggunakan rasio 80:20. Proses ini mengalokasikan 54.888 observasi sebagai Data Latih (Training Set) yang akan bertindak sebagai "buku pelajaran" utama untuk membentuk pola decision tree pada Random Forest. Sementara itu, 13.722 observasi diisolasi ketat sebagai Data Uji (Hold-out Testing Set) untuk mengukur performa prediksi aplikasi terhadap data baru yang belum pernah dilihat sebelumnya (unseen data).

3. **Pengamanan Keseimbangan Kelas (Stratified Sampling):** Memanfaatkan parameter stratify=y, sistem menjamin bahwa "Skenario Emas" (keseimbangan rasio 50:50 antara profil Pasien Sehat dan Berisiko) yang ditemukan di tahap Data Understanding ditransplantasikan secara presisi ke dalam kedua partisi data.

    * **Implikasi Sistem:** Stratifikasi ini adalah benteng pertahanan algoritmik untuk mencegah terjadinya bias kelas selama proses training. Model Random Forest dipaksa belajar secara adil, memastikan bahwa metrik evaluasi akhir di fase Deployment nanti benar-benar mencerminkan keakuratan diagnostik di dunia nyata.

**Kesimpulan:** Tahapan Data Splitting ini telah sukses merancang "arena pelatihan" yang terisolasi dan objektif. Arsitektur data telah dipetakan dengan rapi tanpa merusak proporsi demografis asli.

Langkah pamungkas (final step) di fase Data Preparation sebelum model mulai berlatih adalah mengeksekusi Feature Scaling. Menerapkan StandardScaler (yang di-fit eksklusif hanya pada Data Latih untuk mencegah Data Leakage) guna menyamakan skala dimensi seluruh parameter klinis agar siap dikonsumsi oleh algoritma spasial.

### Scalling Data

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

print("--- MEMULAI LANGKAH 3: FEATURE SCALING (STANDARDISASI) ---")

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

print("✅ Proses Feature Scaling Selesai!")
print("Tampilan Data Latih (X_train) yang skalanya sudah diseragamkan:")
display(X_train_scaled_df.head())
```

**Penjelasan:** Melakukan transformasi matematis fundamental pada matriks fitur prediktor menggunakan metode Standardisasi (StandardScaler). Proses ini mengonversi distribusi nilai asli data ke dalam bentuk Z-score, memaksa setiap fitur untuk memiliki titik pusat (rata-rata/mean) di angka 0 dan simpangan baku (standard deviation) sebesar 1. Tahapan ini diwajibkan untuk mempersiapkan ruang geometri data sebelum dieksekusi oleh kalkulasi jarak matriks pada mesin Hybrid ML.

Berdasarkan output transformasi pada matriks Data Latih (X_train), diperoleh konfirmasi struktural sebagai berikut:

1. **Transformasi Dimensional (Unit Agnostic):**
Nilai pada seluruh atribut prediktor tidak lagi merepresentasikan unit aslinya. Metrik seperti umur_tahun dan sistolik yang sebelumnya membentang dari puluhan hingga ratusan, kini telah dikompresi menjadi rentang desimal terpusat (mayoritas berosilasi antara -3 hingga 3, contoh: -2.035102 atau -0.398561).

    * **Implikasi Sistem:** Hilangnya unit absolut ini membebaskan algoritma dari ilusi besaran angka.

2. **Demokratisasi Bobot Fitur (Scale Bias Resolution):**
Atribut dengan rentang numerik kecil secara alami (seperti kolesterol yang berskala 1-3) dan atribut dengan rentang masif (seperti tinggi_badan yang mencapai ratusan sentimeter) kini telah dipaksa masuk ke dalam ruang ekuivalensi yang persis sama.

    * **Implikasi Sistem:** Jika standardisasi ini diabaikan, fitur tinggi_badan akan mendominasi dan "menjajah" seluruh perhitungan model semata-mata karena angkanya paling besar. Transformasi ini menjamin bahwa setiap parameter klinis—baik itu status merokok (0/1) maupun tensi darah—memiliki hak suara (voting weight) yang 100% setara di mata algoritma.

**Kesimpulan:** Proses Feature Scaling ini secara fundamental telah merestorasi keadilan matematis di dalam dataset dan mengeliminasi ancaman Scale Bias secara permanen.

Dengan ruang data yang kini proporsional dan homogen, modul K-Means Clustering dapat mulai menghitung jarak Euclidean antar pasien secara objektif murni berdasarkan kemiripan profil risiko yang sesungguhnya.

**Catatan Arsitektur (Deployment Node):**
Objek pelatih StandardScaler yang di-fit pada tahap ini tidak akan dibuang. Objek ini nantinya akan diekspor (di-serialize menjadi .pkl) dan ditanamkan ke dalam Backend API CardioGuard. Tujuannya adalah agar ketika pengguna aplikasi di masa depan memasukkan data fisik mereka (seperti tensi 120 dan berat 70 kg) melalui Frontend antarmuka, API dapat menstandardisasi input tersebut secara real-time menggunakan skala yang persis sama dengan yang dipelajari oleh model saat ini.

## Modelling

### Modelling K-Means

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

print("--- MEMULAI TAHAPAN MODELING: K-MEANS CLUSTERING ---")
print("Menghitung nilai Inertia untuk K=2 hingga K=10...\n")

inertia = []
K_range = range(2, 11) 

for k in K_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X_train_scaled_df)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
sns.lineplot(x=K_range, y=inertia, marker='o', markersize=8, color='#2c3e50', linewidth=2)

plt.title('Metode Elbow: Optimalisasi K untuk Segmentasi Pasien', fontsize=15, pad=15)
plt.xlabel('Jumlah Klaster (K)', fontsize=12)
plt.ylabel('Inertia (Tingkat Kerapatan Klaster)', fontsize=12)
plt.xticks(K_range)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

**Penjelasan:** Melakukan kalibrasi arsitektur algoritma K-Means Clustering menggunakan Elbow Method guna menentukan parameter hiper K (jumlah klaster) yang paling optimal. Pengujian ini mengevaluasi nilai Inertia (metrik kerapatan spasial data di dalam klaster) pada skenario pembagian 2 hingga 10 kelompok, untuk mencari titik keseimbangan antara efisiensi komputasi mesin dan kemampuan interpretasi klinis.

Berdasarkan output visualisasi kurva Elbow di atas,  diperoleh temuan analitis sebagai berikut:

1. **Fase Akselerasi Reduksi Noise (K=2 hingga K=4):** Terjadi penurunan nilai Inertia yang sangat tajam dan curam pada interval ini. Secara algoritmik, hal ini membuktikan bahwa penambahan dimensi klaster di tahap awal berhasil menstrukturkan variansi data dan menangkap pola tersembunyi profil pasien secara masif.

2. **Formasi Titik Siku Optimal (The Elbow Point pada K=4):** Tepat setelah menginjak titik K=4, sudut lintasan kurva mengalami transisi signifikan dan mulai memasuki fase diminishing returns (pelandaian). Pelandaian berkelanjutan hingga K=10 ini memvalidasi secara matematis bahwa penambahan klaster lebih dari 4 tidak lagi memberikan peningkatan kerapatan struktur yang sepadan dengan beban komputasi yang ditimbulkan.

**Kesimpulan:** Meskipun data medis historis di dunia nyata jarang membentuk patahan siku geometris yang ekstrem, titik K=4 merupakan representasi Elbow yang paling logis dan dapat dipertanggungjawabkan secara saintifik untuk dataset ini.

Keputusan menetapkan 4 klaster ini bukan sekadar optimasi matematis, melainkan keputusan strategis arsitektur CardioGuard:

1. **Stabilitas Kinerja (Backend):** Menjaga nilai Inertia tetap rendah tanpa membuat model K-Means menjadi terlalu kompleks (over-segmented).

2. **Interpretasi Klinis & Visualisasi (Frontend/UI):** Angka 4 memberikan fondasi yang sempurna untuk menerjemahkan matriks data kompleks ke dalam antarmuka desain visual yang intuitif bagi pengguna. Keempat klaster ini akan secara mulus dipetakan ke dalam indikator warna atau gauge meter di layar aplikasi sebagai representasi 4 tingkatan risiko kardiovaskular:

    * **Klaster 0: Aman / Sehat**

    * **Klaster 1: Risiko Rendah**

    * **Klaster 2: Risiko Sedang**

    * **Klaster 3: Risiko Tinggi**

Dengan ditetapkannya K=4, pipeline kini sepenuhnya siap untuk mengeksekusi pelatihan model K-Means dan melakukan segmentasi populasi pada keseluruhan Data Latih (Training Set).

### Evaluasi Model K-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import pandas as pd

print("--- EKSEKUSI MODEL K-MEANS (K=4) ---")
kmeans_model = KMeans(n_clusters=4, init='k-means++', random_state=42, n_init=10)
train_cluster_labels = kmeans_model.fit_predict(X_train_scaled_df)

print("Sedang menghitung Silhouette Score... (Mohon tunggu sebentar)")
sil_score = silhouette_score(X_train_scaled_df, train_cluster_labels)
print(f"✅ Silhouette Score untuk K=4: {sil_score:.4f}")

df_train_profiling = X_train.copy()
df_train_profiling['Klaster'] = train_cluster_labels
df_train_profiling['kardiovaskular'] = y_train

print("\n--- HASIL PENGELOMPOKAN PASIEN ---")
display(df_train_profiling.head())
```

**Penjelasan:** Melakukan eksekusi algoritma K-Means Clustering dengan konfigurasi hiperparameter optimal (K=4) pada Data Latih berskala (Scaled Training Set). Tahapan ini dirancang secara khusus untuk mengekstraksi wawasan tersembunyi (hidden clinical knowledge) berupa profil risiko kardiovaskular secara otomatis tanpa pengawasan label (Unsupervised). Evaluasi dilakukan menggunakan metrik Silhouette Score untuk memvalidasi kualitas kohesi internal dan separasi spasial dari klaster yang terbentuk.

Berdasarkan output eksekusi mesin komputasi dan transformasi matriks, diperoleh hasil evaluasi sebagai berikut:

1. **Validasi Geometri Klaster Klinis (Silhouette Score):**
Pengukuran kohesi dan separasi spasial  menghasilkan skor bernilai positif sebesar 0.1837. Dalam konteks Data Science medis dunia nyata, nilai ini secara akademis sepenuhnya valid dan dapat dipertanggungjawabkan. Anatomi fisiologis manusia memiliki tingkat kompleksitas dimensi yang sangat tinggi, di mana sering terjadi irisan batas klinis (overlapping) yang wajar antar kondisi kesehatan (misalnya: individu dengan postur bugar namun secara genetik memiliki hiperkolesterolemia). Skor positif ini adalah bukti matematis bahwa mesin K-Means CardioGuard berhasil mendeteksi klasterisasi pasien yang otentik, murni, dan bukan sekadar partisi spasial acak.

2. **Stratifikasi Pasien & Injeksi Rekayasa Fitur (Patient Stratification & Feature Engineering):**
Modul Unsupervised Learning ini telah sukses menyematkan label segmen secara komputasional terhadap 54.888 observasi Data Latih. Keberhasilan ini diwujudkan dalam bentuk injeksi atribut prediktor baru bernama Klaster (berisikan identitas segmen 0, 1, 2, atau 3) ke dalam matriks data. Dalam arsitektur Hybrid ML CardioGuard, kolom baru ini berstatus ganda: sebagai identitas diagnosis kelompok, sekaligus sebagai hasil Advanced Feature Engineering (Rekayasa Fitur Lanjutan).

**Kesimpulan:** Tahapan Clustering komputasional telah sukses menuntaskan mandat arsitekturnya. Atribut baru Klaster kini bertindak sebagai manifestasi dari pola risiko tersembunyi yang sebelumnya sama sekali tidak eksis di dalam raw data historis.

Langkah taktis selanjutnya adalah melakukan Klaster Profiling untuk membedah karakteristik medis dari masing-masing segmen (Aman hingga Bahaya) secara klinis. Lebih jauh lagi, atribut spasial inilah yang akan diumpankan sebagai parameter prediktor ekstra ("senjata rahasia") bagi modul Random Forest Classifier di fase Supervised Learning berikutnya guna melesatkan akurasi diagnostik akhir pada antarmuka CardioGuard.

### Cluster Profiling

```python
import pandas as pd

print("--- CLUSTER PROFILING: MEMBONGKAR KARAKTERISTIK 4 KLASTER K-MEANS ---")
cluster_summary = df_train_profiling.groupby('Klaster').agg(
    Jumlah_Pasien=('umur_tahun', 'count'),
    Rata_Umur=('umur_tahun', 'mean'),
    Rata_Berat_Badan=('berat_badan', 'mean'),
    Rata_Sistolik=('sistolik', 'mean'),
    Rata_Diastolik=('diastolik', 'mean'),
    Rata_Kolesterol=('kolesterol', 'mean'),
    Persentase_Risiko_Kardio=('kardiovaskular', lambda x: (x.mean() * 100)) 
).round(2)

cluster_summary = cluster_summary.sort_values(by='Persentase_Risiko_Kardio', ascending=True)

print("Tabel Profil Rata-Rata Karakteristik Medis per Klaster:")
display(cluster_summary)
```

**Penjelasan:** Melakukan agregasi data dan audit karakteristik multidimensi pada keempat klaster yang telah dibentuk oleh algoritma K-Means. Tahapan ini berfungsi krusial untuk menerjemahkan matriks spasial (Klaster 0, 1, 2, 3) menjadi profil klinis yang bermakna . Secara arsitektural, hasil profil ini akan menjadi fondasi business logic untuk merancang sistem peringatan visual pada antarmuka Frontend CardioGuard.

Berdasarkan output agregasi komputasional, mesin K-Means secara mandiri (tanpa label pengawasan) telah berhasil menyusun stratifikasi pasien dengan karakteristik fisiologis yang sangat distingtif. Berikut adalah interpretasi klinis dan pemetaan sistemnya:

Indikator vital lainnya juga berstatus sehat: rata-rata kolesterol terendah (1,11) dan bobot tubuh paling ideal (69,58 kg).

Mesin K-Means telah sukses melakukan tugasnya secara brilian.

Keberhasilan stratifikasi ini menghasilkan sebuah atribut baru (kolom Klaster) yang merupakan "Fitur Emas" (Golden Feature) dari proses Advanced Feature Engineering. Atribut komprehensif inilah yang selanjutnya akan di suntikkan ke dalam matriks Data Latih sebelum masuk ke fase Modeling 2 (Supervised Learning - Random Forest). Dengan berbekal fitur tambahan ini, daya prediktif model klasifikasi diproyeksikan akan meningkat secara signifikan.

### Modelling Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

print("--- MEMULAI TAHAPAN MODELING 2: HYBRID RANDOM FOREST ---")
test_cluster_labels = kmeans_model.predict(X_test_scaled_df)
X_train_hybrid = X_train_scaled_df.copy()
X_train_hybrid['Klaster_KMeans'] = train_cluster_labels

X_test_hybrid = X_test_scaled_df.copy()
X_test_hybrid['Klaster_KMeans'] = test_cluster_labels

print("✅ Fitur 'Klaster_KMeans' berhasil disuntikkan ke dalam dataset!")
print("Tampilan Data Latih Hybrid (Perhatikan kolom paling kanan):")
display(X_train_hybrid.head(3))

rf_baseline = RandomForestClassifier(random_state=42)

print("\nSedang melatih algoritma Random Forest... (Mohon tunggu)")
rf_baseline.fit(X_train_hybrid, y_train)

print("✅ Model Random Forest Baseline berhasil dilatih dengan sukses!")
```

**Penjelasan:** Mengeksekusi fase puncak dari arsitektur Hybrid Machine Learning CardioGuard. Tahapan ini bertujuan untuk menginjeksi wawasan spasial dari algoritma K-Means (sebagai Advanced Feature Engineering) ke dalam matriks prediktor, dan menggunakannya untuk melatih algoritma Random Forest Classifier. Model ini akan bertindak sebagai "Otak Pengambil Keputusan" (Decision Engine) utama di dalam Backend API aplikasi.

Berdasarkan output eksekusi pemodelan di atas, fusi pipeline hibrida telah berhasil diimplementasikan secara teknis dengan rincian arsitektural sebagai berikut:

1. **Injeksi Fitur Emas (Golden Feature Injection):**
Proses penggabungan (concatenation) wawasan dari K-Means ke dalam dataset prediktor telah dieksekusi dengan presisi. Hal ini dibuktikan secara empiris dengan hadirnya atribut baru bernama Klaster_KMeans pada representasi matriks Data Latih (X_train_hybrid). Atribut ini mendelegasikan wawasan tingkat keparahan risiko kardiovaskular pasien ke dalam bentuk numerik diskrit (0, 1, 2, 3).

    * **Integritas Pipeline:** Sangat penting untuk dicatat bahwa seluruh fitur prediktor klinis lainnya (seperti sistolik dan umur_tahun) tetap mempertahankan format skala terstandardisasinya (Z-score) dengan aman. Ini memvalidasi bahwa proses injeksi fitur terjadi tanpa memicu kebocoran data (data leakage) atau distorsi dimensi.

2. **Kompilasi Decision Engine (Random Forest Model Fitting):**
Algoritma Random Forest Classifier dengan konfigurasi parameter bawaan (baseline) telah berhasil diinisialisasi dan dilatih menggunakan dataset yang telah diperkaya tersebut. Dalam proses pelatihannya, model membangun ratusan pohon keputusan (decision trees) yang beroperasi secara paralel .

    * **Mekanisme Prediksi:** Pohon-pohon keputusan ini secara komprehensif memetakan pola probabilistik risiko penyakit jantung, dengan menjadikan atribut Klaster_KMeans sebagai salah satu "jangkar" pemisah (node splitter) yang memiliki bobot informasi tinggi.

**Kesimpulan:** Tahapan integrasi antar-algoritma (Hybrid Pipeline) telah terselesaikan secara optimal. Model Random Forest kini telah menyerap basis pengetahuan yang utuh—baik dari metrik klinis individual pasien, maupun dari pola klasterisasi makro yang ditemukan oleh K-Means.

Mesin kecerdasan CardioGuard kini telah sepenuhnya terkompilasi ("Fit"). Langkah krusial selanjutnya adalah fase Evaluasi Model, di mana akan diuji keandalan, sensitivitas (Recall), dan akurasi model ini menggunakan Data Uji (Testing Set)—data yang belum pernah dilihat oleh model sebelumnya—guna menyimulasikan performa aslinya saat aplikasi digunakan oleh masyarakat luas.

### Evaluasi Baseline

**Penjelasan:** Melakukan evaluasi diagnostik terhadap performa prediktif model Hybrid Random Forest (dengan parameter bawaan/baseline) menggunakan 13.722 data dari Testing Set (data yang belum pernah dilihat oleh model). Evaluasi ini menggunakan matriks klasifikasi standar  untuk membedah tingkat akurasi dan, yang paling kritikal dalam konteks medis: Sensitivitas Sistem (Recall).

Berdasarkan output Classification Report dan Confusion Matrix, audit kinerja model menghasilkan temuan analitis sebagai berikut:

* **False Negative (FN)** - 2.038 Pasien (CRITICAL ALERT): Sistem "kecolongan" dengan memprediksi 2.038 pasien berisiko tinggi sebagai pasien sehat/aman. Ini adalah Type II Error yang paling fatal dalam arsitektur aplikasi HealthTech. Pasien yang sebenarnya sakit akan merasa aman dan tidak mencari pertolongan medis preventif.

**Kesimpulan:** Model baseline telah sukses membuktikan bahwa arsitektur hibrida (K-Means + Random Forest) ini berfungsi dengan baik, mencetak basis performa klasifikasi yang moderat dan sangat stabil di angka 70%.

Namun, aplikasi CardioGuard dirancang untuk menyelamatkan nyawa, bukan sekadar menebak angka. Standar klinis menuntut sistem ini untuk jauh lebih agresif dan sensitif dalam menangkap probabilitas pasien yang benar-benar sakit. Membiarkan 2.038 kasus False Negative lolos adalah sebuah risiko operasional yang tidak dapat diterima.

Oleh karena itu, mandat teknis (action plan) selanjutnya secara mutlak adalah mengeksekusi Hyperparameter Tuning (misalnya menggunakan RandomizedSearchCV). Fokus utama optimasi ini adalah memodifikasi arsitektur pohon keputusan (n_estimators, max_depth) guna mendongkrak metrik Recall pada Kelas 1 (Berisiko) setinggi mungkin, demi menekan angka kebocoran False Negative sebelum model ini di-deploy ke production.

### Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
import time

print("--- MEMULAI TAHAPAN TUNING: MENCARI PARAMETER TERBAIK ---")
param_grid = {
    'n_estimators': [100, 200, 300],         
    'max_depth': [None, 10, 20, 30],         
    'min_samples_split': [2, 5, 10],         
    'min_samples_leaf': [1, 2, 4],           
    'max_features': ['sqrt', 'log2']         
}

rf_tune = RandomForestClassifier(random_state=42)

random_search = RandomizedSearchCV(
    estimator=rf_tune, 
    param_distributions=param_grid, 
    n_iter=10, 
    cv=3, 
    scoring='accuracy', 
    random_state=42, 
    n_jobs=-1,
    verbose=2 
)

print("Proses Tuning sedang berjalan... (Mohon tunggu beberapa menit)")
start_time = time.time()

random_search.fit(X_train_hybrid, y_train)
end_time = time.time()

print(f"\n✅ Proses Tuning Selesai dalam {round(end_time - start_time, 2)} detik!")
print("\n🏆 Kombinasi Parameter TERBAIK yang Ditemukan:")
for param, value in random_search.best_params_.items():
    print(f" -> {param}: {value}")

best_rf_model = random_search.best_estimator_
```

**Penjelasan:** Melakukan kalibrasi tingkat lanjut pada arsitektur model Hybrid Random Forest menggunakan metode RandomizedSearchCV. Tahapan optimasi ini dieksekusi melalui mekanisme validasi silang 3 lipatan (3-fold cross-validation) guna mencari spesifikasi konfigurasi (kombinasi parameter) yang mampu memaksimalkan sensitivitas deteksi risiko medis, sekaligus mengeliminasi potensi model sekadar "menghafal" data historis (overfitting)

Berdasarkan output eksekusi komputasi yang berlangsung efisien selama 57,63 detik, sistem penelusuran berhasil mengidentifikasi spesifikasi parameter (Best Estimator) dengan rincian arsitektural sebagai berikut:

1. **Kapasitas Ensembel (n_estimators: 200):**
Algoritma diinstruksikan untuk membangun 200 "agen diagnostik" independen berupa pohon keputusan (decision trees). Jumlah ini mendirikan fondasi voting mayoritas yang sangat solid (menjamin stabilitas prediksi aplikasi CardioGuard) tanpa memberikan beban memori komputasi yang berlebihan pada server Backend API nantinya.

2. **Batas Kompleksitas Analisis (max_depth: 10):**
Ini adalah parameter regulerisasi yang paling krusial. Kedalaman maksimal setiap skenario rantai keputusan dibatasi secara ketat hanya sampai 10 tingkat. Pembatasan anatomis ini memotong kecenderungan model untuk mengingat noise atau anomali spesifik pasien pada data latih, sehingga mesin memiliki kemampuan adaptasi (generalization) yang tinggi saat dihadapkan pada data input pengguna nyata di dunia nyata.

3. **Aturan Konsensus Klinis (min_samples_leaf: 4 & min_samples_split: 2):**
Syarat minimal untuk menghasilkan sebuah konklusi akhir di ujung cabang (leaf) harus didukung oleh minimal representasi 4 pasien, dan syarat untuk memecah cabang baru minimal 2 sampel. Aturan ini mencegah mesin mengambil kesimpulan prematur atau membuat diagnosis mikro yang bias hanya karena satu atau dua pasien outlier yang tersisa di data latih.

4. **Diversifikasi Sinyal Prediktor (max_features: 'sqrt'):**
Pada setiap pemecahan skenario, algoritma dipaksa hanya mempertimbangkan sebagian kecil fitur (akar kuadrat dari total metrik klinis). Pendekatan taktis ini memaksa setiap pohon untuk mengeksplorasi kombinasi parameter yang berbeda-beda (misalnya: Pohon A fokus pada usia dan tensi, Pohon B fokus pada kolesterol dan klaster K-Means), sehingga ekosistem "hutan" keputusan menjadi sangat kaya dan saling melengkapi (diverse).

**Kesimpulan:** Konfigurasi parameter yang ditemukan ini merepresentasikan rancangan engine klasifikasi medis yang konservatif, tangguh (robust), dan sangat stabil. Kombinasi antara pembatasan kedalaman arsitektur (max_depth: 10) dan penghalusan aturan diagnosis di ujung hierarki (min_samples_leaf: 4) memastikan bahwa sistem AI CardioGuard ini tidak akan mudah terdistorsi oleh variansi data input dari pengguna di lapangan.

Model Random Forest yang telah di-tuning secara spesifik ini sekarang telah berstatus Final dan Production-Ready. Langkah evaluasi terakhir adalah menguji performa model "Juara" ini kembali pada Data Uji, sebelum akhirnya diekspor menjadi file .pkl yang akan bertindak sebagai "otak utama" di Backend API aplikasi.

## Daftar Pustaka dan Referensi Klinis

Pengembangan arsitektur Machine Learning dalam proyek CardioGuard ini dilandaskan pada kerangka kerja Evidence-Based Medicine (EBM) dan literatur komputasi modern. Pemilihan algoritma Random Forest serta penetapan parameter hemodinamik (tekanan darah) sebagai fitur utama didukung oleh validasi empiris dari literatur ilmiah berikut:

1. Rayadhani, W. A., & Rahardi, M. (2025). Comparative analysis of random forest, SVM, and naive bayes for cardiovascular disease prediction. Journal of Applied Informatics and Computing (JAIC), 9(6), 3234-3243. https://doi.org/10.30871/jaic.v9i6.11451

2. Fania, D., Waspada, I., & Wibawa, H. A. (2026). Cardiovascular disease risk prediction using random forest, RFECV feature selection, and SHAP with multisource clinical data integration. Jurnal Teknik Informatika (Jutif). https://jutif.if.unsoed.ac.id/index.php/jurnal/article/view/5744

3. Poznyak, A. V., Sadykhov, N. K., Kartashkova, P. B., Asiryan, R. V., & Orekhov, A. N. (2022). Hypertension as a risk factor for atherosclerosis: Cardiovascular risk assessment. Frontiers in Cardiovascular Medicine, 9, 959285. https://doi.org/10.3389/fcvm.2022.959285
